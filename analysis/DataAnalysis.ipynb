```python
import pandas as pd
from matplotlib import pyplot as plt
import os
from os.path import dirname
import fnmatch
import random
%matplotlib notebook
```

```python
data_path = ["/Users/quickstep/Documents/DeepLearning/Project/train_data"]
```

```python
angles = []
for train_data_folder in data_path:
    for root, dirnames, filenames in os.walk(train_data_folder):
        for filename in fnmatch.filter(filenames, '*.csv'):
            data_file = os.path.join(root, filename)
            #print (data_file)
            data= pd.read_csv(data_file, header= None)
            data.columns =["center_cam_image","left_cam_image","right_cam_image",\
                      "angle","forward_throttle","reverse_throttle","speed"]

            steering_angle =  data["angle"].values
            angles+=steering_angle.tolist()
```

```python
fig = plt.figure()
ax = plt.axes()
plt.hist(angles)
ax.set_xlabel("Steering Angle")
ax.set_ylabel("Frequency")
ax.grid()
```

```python
### Trying to make the steering angles uniform
```

```python
angles = []
for train_data_folder in data_path:
    for root, dirnames, filenames in os.walk(train_data_folder):
        for filename in fnmatch.filter(filenames, '*.csv'):
            data_file = os.path.join(root, filename)
            #print (data_file)
            data= pd.read_csv(data_file, header= None)
            data.columns =["center_cam_image","left_cam_image","right_cam_image",\
                      "angle","forward_throttle","reverse_throttle","speed"]

            steering_angle =  data["angle"].values
            for angle in steering_angle.tolist():
                if angle ==0.0:
                    if random.uniform(0, 1)>0.7:
                        angles.append(angle)
                else:
                    angles.append(angle)
```

```python
fig2 = plt.figure()
ax2 = plt.axes()
plt.hist(angles)
ax2.set_xlabel("Steering Angle")
ax2.set_ylabel("Frequency")
ax2.grid()
```
